2.畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?


```python
import pandas as pd
import numpy as np
```

```python
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?
boston_df = pd.read_csv('./D009_data/boston.csv')
df_header = boston_df.head(0)

# boston的箱型圖
boston_df.boxplot(figsize=(12,6))

header_index = 0
for x_median in np.median(boston_df, axis=0):
    if x_median > 300 and x_median < 400:
        print("判斷哪個欄位的中位數在300~400之間? ", list(df_header)[header_index])
        print(list(df_header)[header_index], "的中位數：\t", x_median)
    header_index += 1
```

```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
boston_df.plot.scatter(x='NOX', y='DIS', figsize=(12,6))
print('NOX 和 DIS 兩者是負相關')
```
